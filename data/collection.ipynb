```python
import yfinance as yf
import pandas as pd
import os
print(os.getcwd())
```

```python
TICKERS = [
    "AAPL",
    "MSFT",
    "AMZN",
    "TSLA",
    "NVDA",
    "META",
    "SPY"
]

df_raw = yf.download(
    tickers=TICKERS,
    period="730d",      # last 730 calendar days
    interval="1h",      # 1-hour bars
    auto_adjust=False,  # keep raw OHLCV
    threads=True
)

df_raw.to_csv("raw/raw_OHLCV_730d_1hr.csv") #type: ignore
df_raw
```
