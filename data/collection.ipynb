```python
import os
import pandas as pd
import time

# =========================================
# Config
# =========================================
tickers = [
    "SPY", #"QQQ", "IWM",                                                            # Core market index ETFs
    #"AAPL", "MSFT", "AMZN", "TSLA", "NVDA", "META", "GOOGL", "JPM", "XOM", "WMT",   # Mega-cap stocks
    #"AMD", "AVGO", "NLFX", "CRM", "SHOP",                                           # High-growth volatile tech stocks
    #"XLK", "XLF", "XLE", "XLY", "XLV",                                              # Sector ETFs
    #"PLTR", "ROKU", "COIN", "TWLO", "UBER"                                          # Emerging volatile tech stocks

]
START_DATE = "2023-11-01"         # overall start (YYYY-MM-DD)
END_DATE   = "2025-11-23"         # overall end   (YYYY-MM-DD)


# =========================================
# Helper: date chunk generator (3-month chunks)
# =========================================
def generate_date_chunks(start: str, end: str, months: int = 3):
    """
    Yield (chunk_start, chunk_end) as pandas Timestamps, in ~`months`-month chunks
    from start to end.
    """
    start_ts = pd.Timestamp(start)
    end_ts   = pd.Timestamp(end)

    cur = start_ts
    while cur < end_ts:
        nxt = cur + pd.DateOffset(months=months)
        if nxt > end_ts:
            nxt = end_ts
        yield cur, nxt
        cur = nxt

# =========================================
# Chunked fetch wrapper
# =========================================
def fetch_polygon_5min_chunked(symbol: str, start: str, end: str, months: int = 3) -> pd.DataFrame:
    """
    Fetch 5-min data for `symbol` between start and end, in `months`-month chunks.
    Assumes `fetch_polygon_5min` is already defined.
    """
    all_chunks = []

    for chunk_start, chunk_end in generate_date_chunks(start, end, months=months):
        s = chunk_start.strftime("%Y-%m-%d")
        e = chunk_end.strftime("%Y-%m-%d")
        print(f"  Fetching {symbol}: {s} → {e}")

        df_chunk = fetch_polygon_5min(symbol, s, e)
        if df_chunk.empty:
            print("    (no data returned for this chunk)")
            continue

        all_chunks.append(df_chunk)
        time.sleep(12)  # be adhere to request rate limits of 5/minute

    if not all_chunks:
        return pd.DataFrame()

    df = pd.concat(all_chunks)

    # Drop any duplicate timestamps just in case chunks overlap
    df = df[~df.index.duplicated(keep="last")].sort_index()

    return df

# =========================================
# Main loop: download & save per-ticker CSVs
# =========================================
for symbol in TICKERS:
    print(f"\n=== Downloading {symbol} ===")
    df_sym = fetch_polygon_5min_chunked(symbol, START_DATE, END_DATE, months=3)

    if df_sym.empty:
        print(f"  No data for {symbol} in range {START_DATE} → {END_DATE}")
        continue

    out_path = f"raw_5min/{symbol}_{START_DATE}_to_{END_DATE}.csv"
    df_sym.to_csv(out_path)
    print(f"  Saved {len(df_sym)} rows to {out_path}")
```

```python
import pandas as pd
import plotly.express as px

# Path to your CSV
csv_path = "raw_5min/SPY_2023-11-01_to_2025-11-23.csv"

# Load data (timestamp index was saved as a column called 'timestamp')
df = pd.read_csv(csv_path, parse_dates=["timestamp"])

# Extract date and count rows per day
df["date"] = df["timestamp"].dt.date
bars_per_day = (
    df.groupby("date")
      .size()
      .reset_index(name="num_bars")
)

# Plot with Plotly
fig = px.bar(
    bars_per_day,
    x="date",
    y="num_bars",
    title="Number of 5-min Bars per Day (SPY)",
    labels={"date": "Date", "num_bars": "Bars per day"},
)

fig.update_layout(xaxis_rangeslider_visible=True)
fig.show()
```
